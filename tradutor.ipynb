```python
import pandas as pd
import os
import difflib
from googletrans import Translator
```

```python
def similarity_test(str1, str2):
    result =  difflib.SequenceMatcher(a=str1.lower(), b=str2.lower())
    return result.ratio()
```

```python
plots = pd.read_csv("./data/oscar_full_plot.csv", sep=';')
plots.set_index("FILENAME_ES", inplace = True)
```

```python
path = './data/plots_es'
path_en = './data/plots_en'
text = ""
list_files = os.listdir(path)

n = 0

translator = Translator()

results = []
for file in list_files:
    article = open(path + "/" + file, 'r')
    content = article.read()

    try:
        article = open(path + "/" + file, 'r')
        content = article.read()
        translated = translator.translate(content, dest='en')
    except FileNotFoundError:
        print("File", file, "not found. Skipping translation.")
    except Exception as e:  # Catch other potential errors
        print("Error reading file:", e)

    file_en = plots.at[file, "FILENAME_EN"]
    article_en = open(path_en + "/" + file_en, 'r')
    content_en = article_en.read()

    print(file)
    print(n)
    # print(str(translated))

    if similarity_test(str(translated), content_en) > 0.8:
        n += 1

print(n)
```
